# WeRateDogs Analysis and Visualization

 Source (https://twitter.com/dog_rates/status/1551963675468222464?s=20&t=EDLVOL9BVnX__l8Xsjm8gA)

WeRateDogs is a famous Twitter page where dogs are given funny ratings with numerator greater than denominator (e.g. 13/10). I carried out a data wrangling project on tweet data from WeRateDogs Twitter page. The focus of the project is to demonstrate mastery of data wrangling process.

Data wrangling is the processing of transforming raw data to useable data for data analysis and machine learning predictions. Real world data usually come in messy and dirty format which must be prepared in cleaned and structured format before it can be used for analysis and for model building.  

Data wrangling procedure can be categorized into three stages; gathering, assessing and cleaning. In this project, the data used was gathered from three different sources. The datasets were visually and programmatically assessed to uncover 8 quality and 2 tidiness issues in the datasets. The datasets where properly cleaned and combined together for analysis.

After data wrangling, the following were investigated.

1. The most used source of tweet for the WeRateDogs dataset

2. The most popular dog stage recorded from the WeRateDogs tweet

3. Relationship between Favorite Count and Retweet Count
